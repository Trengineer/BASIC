```python
import pandas as pd

# Replace 'your_file.csv' with the path to your CSV file
file_path = '../GALS/data/waterbird_1.0_forest2water2/metadata-ProbVLM.csv'

# Load the CSV into a DataFrame
df = pd.read_csv(file_path)
```

```python
df
```

```python
df['Classes'] = df['img_filename'].str.split('/').str[0]
```

```python
# Replace 'your_file.txt' with the path to your text file
file_path2 = '../ProbVLM/src/ds/annotations/cub/trainvalclasses.txt'

df2 = pd.read_csv(file_path2, header=None, names=["Classes"])
```

```python
df_train_split = pd.merge(df, df2, on="Classes", how="inner")
```

```python
df_train_split
```

```python
# Replace 'your_file.txt' with the path to your text file
file_path3 = '../ProbVLM/src/ds/annotations/cub/seen_test_images.txt'

df3 = pd.read_csv(file_path3, header=None, names=["img_filename"])
```

```python
df3
```

```python
df_train_split_final = df_train_split[~df_train_split['img_filename'].isin(df3['img_filename'])]
df_train_split_final
```

```python
df_train_split_final['split'] = 0
df_train_split_final.drop(['Classes'], axis=1, inplace=True)
df_train_split_final
```

```python
df_train_split[df_train_split['img_filename'].str.contains('rgb', case=False, na=False)]
```

```python
df_valid_final = pd.merge(df3,df_train_split, on="img_filename", how="inner")
df_valid_final['split'] = 1
df_valid_final.drop(['Classes'], axis=1, inplace=True)
df_valid_final
```

```python
df_train_split_final.columns.tolist()
df_valid_final = df_valid_final[df_train_split_final.columns.tolist()]
```

```python
df_valid_final
```

```python
# Replace 'your_file.txt' with the path to your text file
file_path4 = '../ProbVLM/src/ds/annotations/cub/testclasses.txt'

df4 = pd.read_csv(file_path4, header=None, names=["Classes"])
df4
```

```python
df_test_split = pd.merge(df, df4, on="Classes", how="inner")
```

```python
df_test_split['split'] = 2
df_test_split.drop(['Classes'], axis=1, inplace=True)
df_test_split
```

```python
metadata_final = pd.concat([df_train_split_final, df_valid_final, df_test_split], axis=0, ignore_index=True)
metadata_final = metadata_final.sort_values(by='img_id', ascending=True, ignore_index=True)
metadata_final
```

```python
metadata_final.to_csv('../GALS/data/waterbird_1.0_forest2water2/metadata_ProbVLM_final.csv', index=False)
```
